```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# load data
data = pd.read_csv("spambase.data", header=None)

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# split data: 75% train, 25% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# train logistic regression
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# 1a
cm = confusion_matrix(y_test, y_pred)
print("\n1a. Confusion Matrix")
print(cm)

# 1b
accuracy = accuracy_score(y_test, y_pred)
error = 1 - accuracy
print("\n1b. Accuracy and Error")
print("Accuracy:", accuracy)
print("Error:", error)

# 1c
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("\n1c. Precision, Recall, F1 Score")
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

```

```python
coefficients = model.coef_[0]

coef_df = pd.DataFrame({
    "Feature_Index": range(len(coefficients)),
    "Coefficient": coefficients,
    "Abs_Coefficient": np.abs(coefficients)
})

print("\nAll coefficients:")
print(coef_df[["Feature_Index", "Coefficient"]])

print("\nTop 10 features contributing most to prediction (largest absolute coefficients):")
print(coef_df.sort_values(by="Abs_Coefficient", ascending=False).head(10)[["Feature_Index", "Coefficient"]])

print("\nTop 10 positively correlated with SPAM:")
print(coef_df.sort_values(by="Coefficient", ascending=False).head(10)[["Feature_Index", "Coefficient"]])

print("\nTop 10 negatively correlated with SPAM:")
print(coef_df.sort_values(by="Coefficient", ascending=True).head(10)[["Feature_Index", "Coefficient"]])

```

```python
probs = model.predict_proba(X_test)[:, 1]
thresholds = [0.25, 0.5, 0.75, 0.9]

for t in thresholds:
    pred_t = (probs >= t).astype(int)
    acc_t = accuracy_score(y_test, pred_t)
    prec_t = precision_score(y_test, pred_t, zero_division=0)
    rec_t = recall_score(y_test, pred_t, zero_division=0)

    print(f"\nThreshold = {t}")
    print("Accuracy:", acc_t)
    print("Precision:", prec_t)
    print("Recall:", rec_t)

```
