```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

# load data
data = pd.read_csv("spambase.data", header=None)

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values



def k_fold_cv(X, y, model, k):
    n = len(X)
    indices = np.arange(n)
    np.random.shuffle(indices)
    fold_size = n // k
    errors = []

    for i in range(k):

        start = i * fold_size
        end = start + fold_size
        val_idx = indices[start:end]
        train_idx = np.concatenate((indices[:start], indices[end:]))
        X_train = X[train_idx]
        y_train = y[train_idx]

        X_val = X[val_idx]
        y_val = y[val_idx]
        model.fit(X_train, y_train)
        pred = model.predict(X_val)
        accuracy = accuracy_score(y_val, pred)
        error = 1 - accuracy
        errors.append(error)
        print(f"Fold {i+1} error:", error)

    avg_error = np.mean(errors)

    return avg_error
```

```python


k_values = [5, 10]

for k in k_values:

    print(f"\nLogistic Regression with k={k}")

    log_model = LogisticRegression(max_iter=10000)

    log_error = k_fold_cv(X, y, log_model, k)

    print("Average validation error:", log_error)


    print(f"\nLDA with k={k}")

    lda_model = LinearDiscriminantAnalysis()

    lda_error = k_fold_cv(X, y, lda_model, k)

    print("Average validation error:", lda_error)
```
